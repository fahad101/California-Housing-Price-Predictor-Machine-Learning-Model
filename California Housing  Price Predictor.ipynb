### import library

```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler


%matplotlib inline
```

### Read DataSet

```python
df = pd.read_csv("housing.csv")
df.head(3)
```

```python
df.shape 
```

```python
df=df.dropna(how="any")
df.shape
```

# Create a Feature Map column

```python
columns=['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value', 'ocean_proximity']

columns_norm = ['housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income']
```

# Create a Data Set

```python
X_Data = df[columns_norm]
```

```python
Y_Data = df["median_house_value"]
```

# Create a Train Test Split

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Data, df["median_house_value"], test_size=0.3,random_state=101)
```

```python
X_Train.head(2)
```

```python
Y_Train.head(2)
```

# Sklearn Pre Processing

```python
scalar = MinMaxScaler()
```

```python
scalar.fit(X_Data)
```

# Create DataFrame

```python
 scalar.transform(X_Train)[0]
```

# Normalize Data Set using Sci-kit Learn Pre processing

```python
df1 = pd.DataFrame(data = scalar.transform(X_Train),columns=columns_norm)
```

# Create Y data

```python
Y_Train.shape
Y_Train.head(2)
```

# Neural Network 

```python
#feature_long = tf.feature_column.numeric_column("longitude")

#feature_lat = tf.feature_column.numeric_column("latitude")


feature_housing_median_age = tf.feature_column.numeric_column("housing_median_age")


feature_total_rooms = tf.feature_column.numeric_column("total_rooms")


feature_total_bedrooms = tf.feature_column.numeric_column("total_bedrooms")

feature_population = tf.feature_column.numeric_column("population")

feature_households= tf.feature_column.numeric_column("households")


feature_median_income = tf.feature_column.numeric_column("median_income")

```

# Creating Feature Column

```python
feature_col = [feature_housing_median_age, feature_total_rooms,
                 feature_total_bedrooms,feature_population, feature_households, feature_median_income]
len(feature_col)
```

```python
input_fun = tf.estimator.inputs.pandas_input_fn(x=X_Train, 
                                               y = Y_Train,
                                               batch_size=20,
                                               num_epochs=5000,
                                               shuffle=True,)
```

```python
model = tf.estimator.DNNRegressor(hidden_units=[6,6,6], feature_columns=feature_col)
```

```python
model.train(input_fn=input_fun, steps=50000)
```

# Test

```python
eval_func = tf.estimator.inputs.pandas_input_fn(x=X_Test, 
                                               y = Y_Test,
                                               batch_size=20,
                                               num_epochs=1,
                                               shuffle=False)
```

```python
pred_gen = model.predict(eval_func)
```

```python
prediction = list(pred_gen)
```

```python
prediction
```

# Network Predicted 

```python
final_prediction = []

for pred in prediction:
    final_prediction.append(pred["predictions"])
y_pred = pd.DataFrame(data = final_prediction,columns=["Y_Pred"])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_Test,final_prediction)*0.5
```

# Neural Network Prediction

```python
Y_Test.to_csv("y_test",sep='\t', index=False)
y_pred.to_csv("y_pred",sep='\t', index=False)

df1 = pd.read_csv("y_test")
df2 = pd.read_csv("y_pred")

df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

bigdata = pd.concat([df1, df2],axis=1)
bigdata.head(10)
```
